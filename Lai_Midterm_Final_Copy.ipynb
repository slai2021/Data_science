```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = np.array([12,9,20,7,2]) #created two arrays and measurement of a circle 
Y = np.array([38,28,63,22,6])
```

```python
plt.plot(X,Y,'g.') #scatter plot of x and y so you can make a prediction of the new diameter, the depedent variable is what you are trying to predict 
```

```python
m_guess = 0 
b_guess = 0 #initial guess for m and b, taking this data to build the model #if the guess for m and b are 0 then the array doesn't mean anything 
```

$y=mx+b$

```python
m_guess * X + b_guess #the list of guesses for Y
```

```python
y_guess = m_guess * X + b_guess #trying to fit a linear model by taking the input number for every number given. Want to know how it compares with the actual data 
```

```python
y_guess - Y #trying to see how well does the model fit the data #subtract the values from real data and guess 
```

```python
(y_guess - Y)**2 #determine the negative numbers into the positive, the square is to get rid of the minus sign 
```

```python
np.sum((y_guess - Y)**2) #don't want negative to cancel out, total amount of the best fit, 
```

```python
loss = np.sum((y_guess - Y)**2) #loss function 
```

```python
best_loss_so_far = 1000000000

m_best = 0
b_best = 0

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100): #linear regression 
        print(m_guess,b_guess) #slope will be less than -100, no more than 100, taking 100 steps 
        y_guess = m_guess * X + b_guess 

        loss = np.sum((y_guess - Y)**2)

    #try to see every possible of m and then go through all values of b 
        if loss < best_loss_so_far: 
            m_best = m_guess
            b_best = b_guess
            best_loss_so_far = loss #searching through each value of m and b and compare the model of the data location to see the difference between the data and model 


print("Found better fit:")
print(m_best,b_best,loss) 
```
