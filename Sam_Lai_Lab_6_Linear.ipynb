```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
```

```python
loss = np.sum((y_guess -y)**2) #shows how wrong you are with the numbers 
```

```python
for m_guess in np.linspace(-10,10,10): #guessing the different values , but together will try different combos of the slope and intercept, will print the loss 
    for b_guess in np.linspace(-10,10,10):

        y_guess = m_guess * x + b_guess

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,m_guess,b_guess)
```

```python
best_loss = 100000 #tries different values of m and b to guess values 
best_m = 0
best_b = 0 

for m_guess in np.linspace(-10,10,20):
    for b_guess in np.linspace(-10,10,20):
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 

            plt.plot(x,y_guess,'r.');
            plt.plot(x,y,'g.');
            plt.show()

            print(loss,m_guess,b_guess)
```
